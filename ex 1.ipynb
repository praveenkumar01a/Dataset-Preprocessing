SCENARIO 3: Housing Dataset 

Investigate a housing dataset for missing features and relationships between house attributes. 
Dataset (Kaggle â€“ Public): 
https://www.kaggle.com/datasets/harishkumardatalab/housing-price-prediction 

IN-LAB TASKS: 

Load the dataset and inspect columns. 

Detect and summarize missing values. 

Visualize correlations between numerical features using scatter plots and heatmaps. 

Interpret trends for price prediction. 

```python
df = pd.read_csv(r"D:\ml\lab\dataset\third\Housing.csv")
df.columns = df.columns.str.strip()

print(df.columns)
print(df.head())
print(df.info())
print(df.isnull().sum())
```

Detect and summarize missing values. 

Visualize correlations between numerical features using scatter plots and heatmaps. 

Interpret trends for price prediction. 

```python
# Total missing values per column
missing_summary = df.isnull().sum().sort_values(ascending=False)
print("Missing values per column:\n", missing_summary)

missing_percent = (df.isnull().sum() / len(df) * 100).sort_values(ascending=False)
missing_df = pd.concat([missing_summary, missing_percent], axis=1, keys=['MissingCount', 'MissingPct'])
print("Missing values with %:\n", missing_df)

# Get numeric columns that exist in the dataset
numeric_cols = df.select_dtypes(include=["int64","float64"]).columns.tolist()
# Use price as target (it exists in this dataset)
target = 'price'
num_features = [col for col in numeric_cols if col != target][:6]

print(f"Using features: {num_features}")

plt.figure(figsize=(15, 10))
for i, col in enumerate(num_features):
    plt.subplot(2, 3, i+1)
    sns.scatterplot(x=df[col], y=df[target])
    plt.title(f"{col} vs {target}")
plt.tight_layout()
plt.show()

# Correlation matrix for numerical columns
corr_matrix = df.select_dtypes(include=["int64","float64"]).corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, linewidths=0.2)
plt.title("Correlation Heatmap â€“ Numerical Features")
plt.show()

# Correlations with the target
price_corr = corr_matrix[target].sort_values(ascending=False)
print(f"Top correlations with {target}:\n", price_corr.head(10))
print(f"\nLowest correlations with {target}:\n", price_corr.tail(10))
```

SCENARIO 4: Banking Customer Data
Analyze banking data to detect missing account information and understand customer demographics.
Dataset (Kaggle â€“ Public): 
https://www.kaggle.com/datasets/ajay1735/customer-personality-analysis
IN-LAB TASKS:
â€¢	Import dataset into Pandas.
â€¢	Examine structure, null values, and column types.
â€¢	Visualize age, income, and spending patterns using bar and box plots.
â€¢	Summarize insights from visualizations.


```python


# Load dataset (update path if necessary)
df = pd.read_csv("D:\\ml\\lab\\dataset\\fourth data\\marketing_campaign.csv", sep="\t")

# Preview shape and first rows
print("Shape:", df.shape)
print(df.head())

# Data types and null count
print("\nInfo:\n")
df.info()
print("\nMissing values per column:\n", df.isnull().sum())
# Create Age column
df["Age"] = 2026 - df["Year_Birth"]

print(df[["Year_Birth", "Age"]].head())
```

```python


plt.figure(figsize=(8, 5))
sns.histplot(df["Age"].dropna(), bins=20, kde=True, color="skyblue")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(x=df["Age"].dropna(), color="lightgreen")
plt.title("Box Plot of Age")
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(df["Income"].dropna(), bins=20, kde=True, color="salmon")
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(x=df["Income"].dropna(), color="orange")
plt.title("Box Plot of Income")
plt.show()

```

```python
spend_cols = [
    "MntWines","MntFruits","MntMeatProducts",
    "MntFishProducts","MntSweetProducts","MntGoldProds"
]

# Bar plot: average spend per category
avg_spend = df[spend_cols].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=avg_spend.index, y=avg_spend.values, palette="viridis")
plt.title("Average Spending per Product Category")
plt.ylabel("Average Amount Spent")
plt.xticks(rotation=45)
plt.show()

# Boxplot for spending distributions
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[spend_cols], orient="h")
plt.title("Spending Distribution Across Categories")
plt.xlabel("Amount Spent")
plt.show()
```
